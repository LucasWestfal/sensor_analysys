## Análise exploratória com o intuito de verificar que os dados estão sendo recolhidos de forma correta

```python
import numpy as np
import pandas as pd
```

Abaixo dados da almofada

```python
raw_data = pd.read_csv('raw_data/recording_2/almofada',  sep=", | -> ", names = ["timestamp", "intern_clock", 7, 6, 8, 5, 3, 1, 2, 4], engine = "python")
cols = ["timestamp", "intern_clock", 1, 2, 3, 4, 5, 6, 7, 8]
raw_data = raw_data[cols]
raw_data.drop('timestamp', inplace=True, axis=1)
raw_data = raw_data.astype({1:"int", 2:"int", 3:"int", 4:"int", 5:"int", 6:"int", 7:"int", 8:"int"})
raw_data
```

```python
from matplotlib import pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(raw_data["intern_clock"]/1000, raw_data[5], s=1, c='b', marker="s", label='5')
ax1.scatter(raw_data["intern_clock"]/1000, raw_data[6], s=1, c='g', marker="s", label='6')
ax1.scatter(raw_data["intern_clock"]/1000, raw_data[7], s=1, c='y', marker="s", label='7')
ax1.scatter(raw_data["intern_clock"]/1000, raw_data[8], s=1, c='r', marker="s", label='8')
plt.legend(loc='upper left')

plt.xlabel("Tempo em segundos")
plt.ylabel("valor do sensor")

plt.axvline(x = 180, color = 'black', label = 'axvline - full height')
plt.axvline(x = 195, color = 'black', label = 'axvline - full height')
plt.axvline(x = 300, color = 'black', label = 'axvline - full height')
plt.axvline(x = 320, color = 'black', label = 'axvline - full height')
plt.axvline(x = 390, color = 'black', label = 'axvline - full height')

plt.savefig("integridade_dados")

plt.show()
```

perto de 400s os dados do sensor 7 deram problema, provavelmente algum fio desconectou. Esse dado terá de ser jogado fora

Abaixo, encontrando constantes de correção de dados e gerando dados úteis.

```python
raw_data.set_index("intern_clock", inplace = True)
com_peso = raw_data.truncate(before=1, after=18000)
tara = raw_data.truncate(before=19500, after=30000)
util = raw_data.truncate(before=32000, after=39000)
```

```python
tara = tara.sum()/tara.shape[0]
```

```python
tara[:][1] = 0
tara[:][2] = 0
tara[:][3] = 0
tara[:][4] = 0
```

```python
data = util.apply(lambda x: x - tara, axis=1)
com_peso = com_peso.apply(lambda x: x - tara, axis=1)
```

```python
com_peso = com_peso.sum()/com_peso.shape[0]
com_peso = com_peso/com_peso[5]

com_peso[:][1] = 1
com_peso[:][2] = 1
com_peso[:][3] = 1
com_peso[:][4] = 1

data = data.apply(lambda x: x/com_peso, axis=1)
```

```python
data.reset_index(inplace = True)
data
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(data["intern_clock"]/1000, data[5], s=1, c='b', marker="s", label='5')
ax1.scatter(data["intern_clock"]/1000, data[6], s=1, c='g', marker="s", label='6')
ax1.scatter(data["intern_clock"]/1000, data[7], s=1, c='y', marker="s", label='7')
ax1.scatter(data["intern_clock"]/1000, data[8], s=1, c='r', marker="s", label='8')
plt.legend(loc='upper left')

plt.xlabel("Tempo em segundos")
plt.ylabel("valor do sensor")

# plt.savefig("dados corrigidos")

plt.show()
```

```python
data.to_csv('data/almofada.csv', index = False)
```
