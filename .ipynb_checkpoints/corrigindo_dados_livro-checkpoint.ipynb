Abaixo, encontrando constantes de correção de dados e gerando dados úteis.

```python
raw_data.set_index("intern_clock", inplace = True)
com_peso = raw_data.truncate(before=1, after=18000)
aux = com_peso
tara = raw_data.truncate(before=19500, after=30000)
util = raw_data.truncate(before=31000, after=100000)
```

```python
tara = tara.sum()/tara.shape[0]
```

```python
tara[:][1] = 0
tara[:][2] = 0
tara[:][3] = 0
tara[:][4] = 0
```

```python
data = util.apply(lambda x: x - tara, axis=1)
com_peso = com_peso.apply(lambda x: x - tara, axis=1)
```

```python
com_peso = com_peso.sum()/com_peso.shape[0]
com_peso = com_peso/com_peso[5]

com_peso[:][1] = 1
com_peso[:][2] = 1
com_peso[:][3] = 1
com_peso[:][4] = 1

# data = data.apply(lambda x: x/com_peso, axis=1)
```

Agora falta calcular o centro de massa pra usar prever com o modelo

```python
x = [0]*data.shape[0]
y = [0]*data.shape[0]

P1 = [-90, 120]
P2 = [90, 120]
P3 = [-90, -120]
P4 = [90, -120]

for i in range(data.shape[0]):
    
    x[i] = data[5][i]*P1[0] + data[6][i]*P2[0] + data[7][i]*P3[0] + data[8][i]*P4[0]
    x[i] = x[i]/(data[5][i] + data[6][i] + data[7][i] + data[8][i])
    
    y[i] = data[5][i]*P1[1] + data[6][i]*P2[1] + data[7][i]*P3[1] + data[8][i]*P4[1]
    y[i] = y[i]/(data[5][i] + data[6][i] + data[7][i] + data[8][i])
    
data["x"] = x
data["y"] = y
```

```python
data.drop(5, inplace=True, axis=1)
data.drop(6, inplace=True, axis=1)
data.drop(7, inplace=True, axis=1)
data.drop(8, inplace=True, axis=1)
```
